# Aggregating and calculating means

```python
#creating subset
df = ords_prods_merge[:1000000]
```

# Creating loyalty flag

```python
#create a new column with the max orders for each user
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#checking result
ords_prods_merge.head(15)
```

```python
#creation of loyalty flag
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#checking result
ords_prods_merge.head(1)
```

```python
#Frequency of loyalty flag
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

### Observation:

It is my understanding that the frequency check doesn't indicate the number of new/regular/loyal customers: it indicates, for each individual product in each individual order, if the product has been ordered by a new/regular/loyal customer. Which is why naturally the frequency of "new customer" is much lower. I think it would be interested to transform the data frame, grouping by user_id and only displaying the loyalty flag once for each user_id: then we could know how many customers are new/reg/loyal.

# Creating spending flag

```python
#create a new column with the average price
ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#check result
ords_prods_merge.head(5)
```

```python
#creation of spending flag
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spending_flag'] = 'Low spender'
ords_prods_merge.loc[(ords_prods_merge['avg_price'] >= 10), 'spending_flag'] = 'High spender'
```

```python
#check result
ords_prods_merge.head(5)
```

```python
#Frequency of spending flag
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

### Observation:

There are a lot more low spenders than high spenders (or, most of the products ever bought have been bought by low spenders rather than high spenders)

# Creating order frequency flag

```python
#create a new column with the median number of days between orders
ords_prods_merge['median_dspo'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#check result
ords_prods_merge.head(5)
```

```python
#creation of order frequency flag
ords_prods_merge.loc[ords_prods_merge['median_dspo'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_dspo'] <= 20) & (ords_prods_merge['median_dspo'] > 10), 'order_frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_dspo'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
#check result
ords_prods_merge.head(5)
```

```python
#Frequency of order frequency flag
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

"NaN" value accounts for customers that placed only one order.
